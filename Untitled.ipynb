```python
from sqlalchemy import create_engine
import pandas as pd
from sqlalchemy import text
from sentence_transformers import SentenceTransformer
```

```python
import numpy as np
```

```python

file_path = 'train_data.xlsx'

df = pd.read_excel(file_path, header=1)

display(df.head())
```

```python
engine = create_engine('sqlite:///my_database.db')
```

```python


df.to_sql('my_table', con=engine, if_exists='replace', index=False)

print("Данные успешно загружены в базу данных.")
```

```python
query = "SELECT * FROM my_table LIMIT 5"
df = pd.read_sql(query, con=engine)

# Вывод первых 5 строк данных
display(df)
```

```python
with engine.connect() as connection:
    connection.execute(text("ALTER TABLE my_table ADD COLUMN `ответ был вам полезен ?` TEXT"))

df = pd.read_sql('my_table', con=engine)
```

```python
display(df.head())
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
descriptions = df['Решение'].tolist()
vectors_ans = model.encode(descriptions)
```

```python
vectors_str_ans = [np.array2string(vector, separator=',') for vector in vectors_ans]
```

```python
df['синтаксический вектор решения'] = vectors_str_ans
```

```python
descriptions = df['Описание'].tolist()
vectors_qw = model.encode(descriptions)
```

```python
vectors_str_qw = [np.array2string(vector, separator=',') for vector in vectors_qw]
```

```python
df['синтаксический вектор описания'] = vectors_str_qw
```

```python
display(df.head())
```

```python
def parse_vector(vector_str):
    return np.fromstring(vector_str.strip('[]'), sep=',')
```

```python
vectors_ans = df['синтаксический вектор решения'].apply(parse_vector)
```

```python
df['суммированный вектор решения'] = vectors_ans.apply(lambda vec: np.sum(vec, axis=0))
```

```python
vectors_qw = df['синтаксический вектор описания'].apply(parse_vector)
```

```python
df['суммированный вектор описания'] = vectors_qw.apply(lambda vec: np.sum(vec, axis=0))
```

```python
print(df[['суммированный вектор решения', 'суммированный вектор описания']].head(20))
```

```python
df_excel = pd.read_excel('pdf_data.xlsx')

with engine.connect() as connection:
    create_table_query = '''
        CREATE TABLE IF NOT EXISTS pdf_data (
            {}
        )
    '''.format(', '.join([f'{column} TEXT' for column in df_excel.columns.tolist()]))

    connection.execute(text(create_table_query))
    
    df_excel.to_sql('pdf_data', con=engine, if_exists='replace', index=False)

print("Table 'pdf_data' successfully created and data inserted.") 
```

```python
query = "SELECT * FROM pdf_data"
df_pdf_data = pd.read_sql(query, con=engine)

# Вывод первых 5 строк данных
display(df_pdf_data.head(10))
```

```python
descriptions = df_pdf_data['Content'].tolist()
vectors = model.encode(descriptions)
```

```python
vectors_str = [np.array2string(vector, separator=',') for vector in vectors]
```

```python
df_pdf_data['синтаксический вектор'] = vectors_str
```

```python
vectors = df_pdf_data['синтаксический вектор'].apply(parse_vector)
```

```python
df_pdf_data['суммированный вектор'] = vectors.apply(lambda vec: np.sum(vec, axis=0))
```

```python
display(df_pdf_data.head())
```

```python
df_pdf_data.info()
```

```python
df['index'] = range(len(df))
df_pdf_data['index'] = range(len(df_pdf_data))
```

```python
percent_difference = 100 - abs(( df_pdf_data['суммированный вектор'] - df['суммированный вектор решения'] ) / df_pdf_data['суммированный вектор']) * 100
```

```python
filtered_indices = percent_difference[percent_difference >= 70].index
```

```python
filtered_df = df.loc[filtered_indices]
filtered_df_pdf_data = df_pdf_data.loc[filtered_indices]

print("Строки из df:")
display(filtered_df)
print("\nСтроки из df_pdf_data:")
display(filtered_df_pdf_data)
```
